```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low','Medium','High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins = bins, labels = labels , right = False)
```

```python
# Get counts 
counts = df['Income Band'].value_counts()

# Plot bar chart
plt.bar(counts.index.astype(str), counts.values)  # Make sure index is string for labels
plt.xlabel('Income Band')
plt.ylabel('Count')
plt.title('Income Band Distribution')
plt.show()
```

Bivariate Analysis --> goal is to find relationships, associations, or patterns between those two variables.

```python
# hue -> menaing within each bar(predictor column), count the further split by different values of Nationality
for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='Nationality')
```

```python
# Histplot of value count for different Occupation

# --> we are plotting the Histgram plot  to shows the distribution of numerical variable by grouping into column name and counting how many values fall into the column

plt.figure(figsize=(10,15)) #width and height
for i, col in enumerate(categorical_cols):
    if col == 'Occupation':
        continue
    plt.subplot(4,3,i+1)
    sns.histplot(df[col], discrete=True)
    plt.title('Histogram of Occupation Count')
    plt.xlabel(col)
    plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

Numerical Analysis

```python
numerical_cols = ['Fee Structure','Age','Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

# Visualization
plt.figure(figsize=(15,10))
for i, col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    # subplot -> to show multiple plots in the same plot
    sns.histplot(df[col],kde = True, color='coral') 
    #kde -> kernel density Estimate curve -> a smooth curve representing the probability desnity function of the data
    plt.title(col)
plt.tight_layout()
plt.show()

```

Heatmap --> uses colors to show the magnitude of values in a matrix or table, how stronly two are repalted to each other 

```python
numerical_cols = ['Age','Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']


# calcualte person correlation coefficient between evey pair of numerical column in numrical_col
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot =True, cmap = 'viridis', fmt=".2f")
# annot = True -> annotation -> mean seach cell will display the correaltion number
# cmap ="crest" -> is the color pallter used to shade the cell from low to high corelation values
# fmt = ".2mf" -> formats the annotation numbers to 2 decimal places
plt.title("Correlation Matrix HeatMap")
plt.show()
```

### Insights of EDA
##### 1. Deposits and Saving Behavior:
The strongest positive correlation occur among "Bank Deposits" with "Checking Accounts", "Saving Accounts" and "Foreign Currency Account" indicating that customers who maintain high balances in one account type often hold substantial amount/funds across other accounts as well.

##### 2. Income, Age and Accumulation: 
Moderate correlations of Age and Estimated Income with various balances (Superannuation, Savings, Checking) reflect a common financial lifecycle trend: higher income earners and older individuals often accumulate more savings, retirement funds, and may carry higher credit card balances or loans.

##### 3. Low Correaltion with Properties Owned :
Property ownership may depend on external factors (location, real estate market conditions, inheritance, etc.) that are not captured by these particular banking variables. Hence, we see weaker correlations here.
